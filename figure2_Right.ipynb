```python
import numpy as np
import scipy.integrate as spi
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Paradoxical TCS (with reaction r'1) ODEs
# v is the rate of r'1. 
# When it is zero, the system is robust

def paradox1(u, t, fc, v):
    # Unpack the state variables
    x, xp, y, yp, pr, pra = u
    
    # Parameters
    v1 = fc * 25.36580
    v2 = 21.37180
    v3 = 12.950
    v4 = 10000.0
    v5 = 1000.0
    v6 = 0.713820
    v7 = 0.827270
    v8 = 0.00010
    v9 = 0.00010

    # Equations
    x_dot = -v1 * x + v * xp + v2 * xp * y + v6 * pra - v8 * x
    xp_dot = v1 * x - v2 * xp * y - v * xp
    y_dot = -v2 * xp * y + v3 * x * yp + v7 * pra - v9 * y
    yp_dot = v2 * xp * y - v3 * x * yp - v4 * yp * pr + v5 * pra
    pr_dot = -v4 * yp * pr + v5 * pra
    pra_dot = v4 * yp * pr - v5 * pra

    return [x_dot, xp_dot, y_dot, yp_dot, pr_dot, pra_dot]

# Nullcline of yp, i.e yp = f1(ytot)
def f1(x,fc,v):
    v1 = fc*25.36580
    v2 = 21.37180
    v3 = 12.950
    a = v2*v3
    b1 = v2*v3
    b0 = v*v3+v2*v1
    c = v2*v1
    
    res = (b1 * x + b0 - np.sqrt(b1**2 * x**2 + b0**2 + 2 * x * a * (b0 - 2 * c))) / (2 * a)

    return res

# Nullcline of ytot, i.e. ytot = hY(yp) such that yp = hY^{-1}(ytot)
def hY(x):
    v4 = 10000.0
    v5 = 1000.0
    v7 = 0.827270
    v9 = 0.00010
    pr_tot = 0.0166
    p = v7*pr_tot/v9+v5/v4
    q = v5/v4
    return (x - p + np.sqrt(x**2 + p**2 + x * (4 * q - 2 * p))) / 2

```

```python
# Initial conditions
x_0 = 0.220
xp_0 = 4e-0080
y_0 = 5
yp_0 = 6e-0080
pr_0 = 0.01660
pra_0 = 0.0
z_0 = 0.220

# Define initial conditions
x_init_paradox = [x_0,xp_0,y_0,yp_0,pr_0,pra_0]

# Final time - define time span
tf = 10**6

# Fold change applied to v1, the rate of x autophosphorylation
fc = [1, 0.5, 0.1]  # fc = 1 corresponds to growth-limiting conditions

# Non-robust paradoxical TCS endowed with reaction r'1
v = 0 
```

```python
color1 = [0, 0.4470, 0.7410]
color2 = [0.3410, 0.7690, 0.3470]
color3 = [0.8500, 0.3250, 0.0980]
colors = [color1, color2, color3]

v1 = 25.3658
v3 = 12.950
fc_value = 1

# Define ytot1 and ytot2
ytot1 = np.linspace(0, v1 / v3, 1000000)
ytot2 = np.linspace(v1 / v3, 10000, 1000000)
yp2 = v1*fc_value / v3 * np.ones_like(ytot2)

#ytot_values = [130.5,115,90] 

t = np.linspace(0, tf, 1000000)
for fc_value,color in zip(fc,colors):
    # Plot ytot1 and ytot2
    plt.plot(ytot1, fc_value*ytot1, '-', color=color, label=fc)  # yp1 vs ytot1
    plt.plot(ytot2, fc_value*yp2, '-', color=color, label='_nolegend_')  # yp2 vs ytot2
    
    # Solve the ODE system
    u =spi.odeint(paradox1, x_init_paradox, t, args=(fc_value, v))
    y = u[:, 2]
    yp = u[:, 3]
    plt.plot(y[-1] + yp[-1], fc_value*v1/v3, 'o', markeredgecolor=color, label='_nolegend_', markersize=10,fillstyle='none')

# The nullcline hY
ytot_new = np.linspace(y_0,1000,1000000)
yp2 = hY(ytot_new)
plt.plot(ytot_new,yp2,'-', linewidth=2, color='black')

plt.title('Positive autoregulation', fontsize="14")
# plt.legend(['fc = 1','fc = 0.5','fc = 0.1'])
plt.ylabel('${PhoB}_p$', fontsize="14")
plt.xlabel('${PhoB}_{TOT}$', fontsize="14")
plt.xscale('log')
plt.xlim([0.001,200])
plt.ylim([0,2.5])
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
# plt.savefig('figure2Right.png')
plt.show()
```
